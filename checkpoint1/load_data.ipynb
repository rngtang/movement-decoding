```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Delaunay
```

### Loading the data

```python
# Load the data as numpy arrays
img_1 = pd.read_csv('./../data/feaSubEImg_1.csv', header=None).to_numpy()
img_2 = pd.read_csv('./../data/feaSubEImg_2.csv', header=None).to_numpy()
overt_1 = pd.read_csv('./../data/feaSubEOvert_1.csv', header=None).to_numpy()
overt_2 = pd.read_csv('./../data/feaSubEOvert_2.csv', header=None).to_numpy()

# Load the sensor coords
sensor_coords = pd.read_csv('./../data/BCIsensor_xy.csv', header=None).to_numpy()
```

### Plotting Functions

```python
# Function to plot the imagined and overt features
def plot_trials(data, num_trials=5, title="EEG Trial Visualization"):
    # Transpose to shape (num_trials, 204) for trial-wise access
    X = data.T
    
    plt.figure(figsize=(10, 6))
    for i in range(min(num_trials, X.shape[0])):
        plt.plot(X[i], label=f'Trial {i+1}')
    
    plt.xlabel("Feature Index (0 to 203)")
    plt.ylabel("EEG Feature Value")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
def plot_electrode_positions(sensor_coords, title="BCI Electrode Locations"):
    x, y = sensor_coords[:, 0], sensor_coords[:, 1]

    plt.figure(figsize=(6, 6))

    # Connect electrodes in order using a line
    plt.plot(x, y, color='blue', linewidth=1)

    # Draw the points
    plt.scatter(x, y, color='blue', zorder=5)

    # Label selected electrodes (1-based indexing)
    for i, (xi, yi) in enumerate(zip(x, y)):
        if i in {0, 53, 54, 101}:
            plt.text(xi, yi, str(i + 1), fontsize=9, ha='center', va='center')

    plt.title(title)
    plt.xlabel("X Coordinate")
    plt.ylabel("Y Coordinate")
    plt.axis('equal')
    plt.tight_layout()
    plt.show()
```

### Initial plotting for the Sensor data

```python
plot_trials(img_1, title="Imagined Movement - Class 0")
plot_trials(img_2, title="Imagined Movement - Class 1")
plot_trials(overt_1, title="Overt Movement - Class 0")
plot_trials(overt_2, title="Overt Movement - Class 1")
```

### Plotting the sensors

```python
plot_electrode_positions(sensor_coords)
```
